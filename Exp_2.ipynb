```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
data = pd.read_csv('surgeryfile.csv')
data.head()
```

```python

train_labels = data['complication']
train_samples = data.drop('complication',axis=1)

train_samples = preprocessing.normalize(train_samples)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_samples, train_labels, random_state=5)

train_labels = np.array(y_train)
train_samples = np.array(X_train)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import binary_crossentropy
```

```python
model = Sequential()
model.add(Dense(10, activation='sigmoid', input_shape=(24,)))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1, activation= 'sigmoid'))

# model = Sequential([
#     Dense(16, input_shape=(24,),activation ='relu'),
#     Dense(32, activation = 'relu'),
#     Dense(2,  activation = 'softmax')
# ])
```

```python
# model.summary()
```

```python
model.compile(Adam(lr=0.01),loss='binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(train_samples, train_labels, validation_split = 0.1, batch_size = 10, epochs = 100, shuffle = True, verbose =2)
```
